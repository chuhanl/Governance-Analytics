Step 1. Cleaning the first table (main table): life expectancy

```python
import pandas as pd

healthLink="https://en.wikipedia.org/wiki/List_of_countries_by_life_expectancy"

# fetching the tables
healthdata=pd.read_html(healthLink,header=0,flavor="bs4")
healthVars=healthdata[0].copy()

# first rows
healthVars.head()
```

```python
#column positions to drop
whichToDrop=[2,3,4,7,8,9,11,12,13,14,15,16,17,18]

#updating the data frame
healthVars.drop(labels=healthVars.columns[whichToDrop],axis=1,inplace=True)
```

```python
newColNames=['LifeExpectancyatBirth','GenderLifeGapatBirth','LifeExpectancyat60','GenderLifeGapat60']
namesToBeAltered=healthVars.columns[1::]
namesToBeAltered
```

```python
dictForChanges=dict(zip(namesToBeAltered, newColNames))
dictForChanges
```

```python
healthVars.rename(columns=dictForChanges,inplace=True)
```

```python
healthVars
```

```python
#row positions to drop
whichRsToDrop=[0,1]

healthVars.drop(labels=whichRsToDrop,axis=0,inplace=True)
healthVars
```

```python
# resetting indexes
healthVars.reset_index(drop=True,inplace=True)
healthVars
```

```python
#save the file
healthVars.to_csv("health_data.csv",index=False)
```

Step 2. Cleaning the second table: urbanization

```python
urbanLink="https://en.wikipedia.org/wiki/Urbanization_by_country"

urbandata=pd.read_html(urbanLink,header=0,flavor="bs4")
urbanVars=urbandata[0].copy()

# first rows
urbanVars.head()
```

```python
#column positions to drop
whichToDrop=[0,3,4,5]

#updating the data frame
urbanVars.drop(labels=urbanVars.columns[whichToDrop],axis=1,inplace=True)
```

```python
#rename the columns
newColNames=['Country','pctUrbanRate']
namesToBeAltered=urbanVars.columns[0::]
dictForChanges=dict(zip(namesToBeAltered, newColNames))
urbanVars.rename(columns=dictForChanges,inplace=True)

#dropping the brackets and the numbers in them
urbanVars['pctUrbanRate']=urbanVars['pctUrbanRate'].str.replace('\\[.*\\]','')
```

```python
#save the file
urbanVars.to_csv("urban_data.csv",index=False)
```

Step 3. Cleaning the third table: total health expenditure per capita

```python
expendLink="https://en.wikipedia.org/wiki/List_of_countries_by_total_health_expenditure_per_capita"

expenddata=pd.read_html(expendLink,header=0,flavor="bs4")
expendVars=expenddata[1].copy()

# first rows
expendVars.head()
```

```python
#column positions to drop
whichToDrop=[1,2]

#updating the data frame
expendVars.drop(labels=expendVars.columns[whichToDrop],axis=1,inplace=True)
```

```python
#rename the columns
newColNames=['Country','HealthExpenditure']
namesToBeAltered=expendVars.columns[0::]
dictForChanges=dict(zip(namesToBeAltered, newColNames))
expendVars.rename(columns=dictForChanges,inplace=True)
```

```python
#get rid of the ' *'s
expendVars['Country']=expendVars['Country'].str.replace('\\s+','')
expendVars['Country']=expendVars['Country'].str.replace('*','')
```

```python
#save the file
expendVars.to_csv("expend_data.csv",index=False)
```

Step 4. Integrating the urbanization table with the main table

```python
interimData=healthVars.merge(urbanVars,how='outer',indicator='True')
```

```python
# explore matched and unmatched counts:
interimData['True'].value_counts()
interimData[interimData['True']=='left_only'].Country
interimData[interimData['True']=='right_only'].Country
```

Because there is no overlap between left-only items and right-only items, I will simply merge only the coincidences.

```python
interimData=healthVars.merge(urbanVars)
```

```python
interimData
```

```python
#dropping the row for world average
interimData=interimData[interimData["Country"].str.contains("World") == False]
```

Step 5. Integrating the expenditure table with the interim table

```python
finData=interimData.merge(expendVars,how='outer',indicator='True')
```

```python
# explore matched and unmatched counts:
finData['True'].value_counts()
finData[finData['True']=='left_only'].Country
finData[finData['True']=='right_only'].Country
```

```python
# The countries unmatched
UnmatchedLeft=finData[finData['True']=='left_only'].Country.to_list()
UnmatchedRight=finData[finData['True']=='right_only'].Country.to_list()
```

```python
# match them!
from thefuzz import process
[(left, process.extractOne(left, UnmatchedRight)) for left in sorted(UnmatchedLeft)]
```

```python
# Create a dictionary and replace
changesRight={process.extractOne(left, UnmatchedRight)[0]:left for left in UnmatchedLeft if left != 'North Korea'}
expendVars.Country.replace(changesRight,inplace=True)
```

```python
# redoing merge
finData=interimData.merge(expendVars)
```

```python
#format the data
finData.dtypes

finData.LifeExpectancyatBirth=finData.LifeExpectancyatBirth.astype(float)
finData.GenderLifeGapatBirth=finData.GenderLifeGapatBirth.astype(float)
finData.LifeExpectancyat60=finData.LifeExpectancyat60.astype(float)
finData.GenderLifeGapat60=finData.GenderLifeGapat60.astype(float)
finData.pctUrbanRate=finData.pctUrbanRate.astype(float)
```

```python
finData.to_csv('final_data.csv',index=False)
```
